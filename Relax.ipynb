```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
tr = h2o.import_file("relax3.csv")
```

```python
y='adopted'
```

```python
splits = tr.split_frame(ratios = [0.8], seed = 1)
train = splits[0]
test = splits[1]
```

```python
x = train.columns
y = "adopted"
x.remove(y)
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
# aml = H2OAutoML(max_runtime_secs = 60, seed = 1, project_name = "CCS")
# aml.train(y = y, training_frame = tr)
aml = H2OAutoML(max_runtime_secs = 90) #include_algos=['DRF']
aml.train(x=x, y=y, training_frame=train)
```
